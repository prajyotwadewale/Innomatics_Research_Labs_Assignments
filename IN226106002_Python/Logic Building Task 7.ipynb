```python
#Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")


# Sample Input
light_name = "Bedroom Light"
action = "ON"

# Create object
light = SmartLight(light_name)

# Perform action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
```

```python
#Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
name = "Rahul"
emp_id = "EMP102"
department = "IT"

# Create object
emp = Employee(name, emp_id, department)

# Display ID card
emp.display_id_card()
```

```python
#Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
name = "Anita"
phone = "9876543210"

# Create object
contact1 = Contact(name, phone)

# Display contact information
contact1.display_contact()
```

```python
#Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
product_name = "Headphones"
price = 2499

# Create object
item = Product(product_name, price)

# Display price tag
item.display_price_tag()
```

```python
#Problem 5: Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie_name = "Inception"
rating = 4.8

# Create object
movie1 = Movie(movie_name, rating)

# Display movie details
movie1.display_movie()
```

```python
#Problem 6: Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
customer_name = "Suman"
address = "Hyderabad"

# Create object
delivery1 = Delivery(customer_name, address)

# Display delivery details
delivery1.display_delivery()
```
