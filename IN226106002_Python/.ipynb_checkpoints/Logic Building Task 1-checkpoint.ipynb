```python
#2. Pass / Fail Analyzer 👇

marks = [45, 78, 90, 33, 60] #marks are given

pass_students = 0
fail_students = 0

for mark in marks:
    if mark >= 50:
        pass_students = pass_students + 1
    else:
        fail_students = fail_students + 1

print("Total Students:", pass_students + fail_students)
print("Passed Students:", pass_students)
print("Failed Students:", fail_students)
```

```python
#3. Simple Data Cleaner 👇

names = [" Alice ", "bob", " CHARLIE "] #given

cleaned_names = []

for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"] #given

for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))
    
    if len(msg) > 10:
        print("Flag: Long Message")
    
    print()  # for spacing
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count = error_count + 1

print("Total Error Count:", error_count)
```
